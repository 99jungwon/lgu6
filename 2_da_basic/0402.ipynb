# loc vs iloc
 - 코드 비교

```python
tips[tips['day'] == 'Sat'].reset_index(drop=True).head(1)
```

```python
# loc
# tips.loc[행, 열]

tips.loc[0:2, ['total_bill', 'tip', 'day']]
```

```python
# iloc
# tips

tips.iloc[0:2, [0, 1, 4]]
```

```python
tips.loc[tips['day'] == 'Sat', ['total_bill', 'tip', 'day']].reset_index(drop = True).head(1)
```

```python
# total_bill 11 이하인것만 조회
tips.loc[tips['total_bill'] <= 11, :].reset_index(drop = True)
```

```python
# time가 dinner인 것만 조회
tips.loc[tips['time'] == "Dinner", :].reset_index(drop = True)
```
