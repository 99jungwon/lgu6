# Group By 연산
 - 집계함수 같이 활용
 - 분석의 목적 : 그룹별로 비교하고 싶을 때 사용

```python
tips.groupby('sex')['tip'].mean()

# 통계적으로 유의한가 : 기초 통계 영역으로 넘어감
```

## 요일별 총 결제 금액 구하기
 - 집계함수 : sum()

```python
tips.groupby('day')['total_bill'].sum()
```

## 요일 & 성별로 팁 평균

```python
result = tips.groupby(['day', 'sex'], observed=True)['tip'].mean()
result.unstack()
```

## 식사 시간별 인원 수 평균
 - time, size

```python
tips.groupby('time', observed=True)['size'].mean()
```

## 집계함수 2개 사용하기

```python
tips.groupby(['day', 'sex'], observed=True)[['tip', 'total_bill']].agg(['sum', 'mean'])
```

```python
result = tips.groupby(['day', 'sex'], observed=True)[['tip', 'total_bill']].agg(['sum', 'mean'])
result.columns = ['_'.join(col) for col in result.columns]
result.reset_index()
```

```python
# 파일 내보내기
result.to_csv("dataset/tip_groupby.csv", index=False)
```

```python
result.to_excel("dataset/tip_groupby.xlsx",
                sheet_name="sheet_name_1",
                index=False)
```
